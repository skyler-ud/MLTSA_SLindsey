```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/fedhere/MLTSA_FBianco/refs/heads/main/data/consumptionUS.csv')
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
x=df['year']
y=df['consumption']
```

```python
import numpy as np
df_nonan=df.dropna()
y=df_nonan['consumption']
x=df_nonan['year']
X=np.c_[np.ones((len(df),1)),x]
lr.fit(X,y)
lr.coef_
```

```python
import matplotlib.pyplot as plt
ax=df_nonan.plot(x,lr.predict(X),kind='scatter')
ax.plot()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(fit_intercept=True)
lr.fit(X,y)
lr.fit(X,y)
lr.coef_
```

```python
import matplotlib.pyplot as plt
ax=df.plot(x='year',y='consumption')
ax.plot(x,lr.predict(X))
```
